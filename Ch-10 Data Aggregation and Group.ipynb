```python
frame= pd.DataFrame({
    'key1':np.random.rand(1000),
    'key2':np.random.rand(1000)
})
frame[:10]
bin = pd.cut(frame.key1,4)
bin.describe()

```

```python
def get_stats(group):
    return {
        'mean':group.mean(),  'count':group.count(),
        'min':group.min(),   'max':group.max()
    }
grouped = frame.key2.groupby(bin)
grouped.mean()
grouped.apply(get_stats)
```

```python
#  return quantile numbers
grouping = pd.qcut(frame.key1,5,labels=False)  #qcut(x, q, labels, retbins, precision, duplicates)
gr = frame.key2.groupby(grouping)
gr.apply(get_stats)
```

```python
suits = ['H', 'S', 'C', 'D']
card_val = (list(range(1, 11)) + [10] * 3) * 4
base_names = ['A'] + list(range(2, 11)) + ['J', 'K', 'Q']
cards = []
for suit in ['H', 'S', 'C', 'D']:    
    cards.extend(str(num) + suit for num in base_names)
deck = pd.Series(card_val, index=cards)
deck
len(cards)
deck[:13]
```

```python
def draw(deck,n=3):
    return deck.sample(n)
draw(deck)
get_suit = lambda card: card[-1]
dk = deck.groupby(get_suit).apply(draw,n=2)
dk
```
